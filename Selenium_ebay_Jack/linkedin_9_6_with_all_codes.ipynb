```python
from datetime import datetime
import time
import numpy as np 


start = datetime.now()

job_data = pd.DataFrame(columns=['job_titles','companies', 'job_links', 'profile_matches','salary','promoted','location'])

total_pages = 3

for i in range(25,total_pages*25+25,25):
      
    
    
    #print(i)
    driver.get('https://www.linkedin.com/jobs/search/?currentJobId=3710744018&f_TPR\
    		=r604800&f_WT=2&geoId=103644278&keywords=data%20scientist&location=\
            United%20States&refresh=true&start='
               +str(i))
    time.sleep(np.random.randint(12, 30)) 
    
    
#    job_listings = driver.find_elements(By.CSS_SELECTOR, ".job-card-container")
    job_listings = driver.find_elements(by="css selector", value=".job-card-container")

    #job_listings = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "job-card-container")))
  
    # Scroll down to load more job listings
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    driver.maximize_window()
   
    # Wait for new listings to load (you may need to adjust the timeout)
    time.sleep(np.random.randint(12, 30)) 
  # You can adjust the sleep duration as needed
    


    for job in job_listings:
        job_html = bs(job.get_attribute('innerHTML'), 'html.parser')
        
        try:
            title = job_html.find(class_='job-card-list__title').text
        except:
            title = ''
        
        #print(title)
        try:
            company = job_html.find(class_='job-card-container__primary-description').text
        except:
            company = ''
        
        
        try:
            link = job_html.a['href']
        except:
            link = ''
        

        # Check if "Your profile matches this job" is present
        try:
            profile_match_element = job_html.find(class_='job-card-container__job-insight-text').text
        except:
            profile_match_element = ''
            
        try:
            location = job_html.find(class_='job-card-container__metadata-item').text
        except:
            location = ''
        
        try:
            # Extract salary-related keywords from the job listing text
            job_text = job_html.get_text()
            salary_keywords = re.findall(r'\$\d+\s*[-–]\s*\$\d+|\$\d+(?:[\s\-–]\d+)?\s*(?:per|\/)?\s*\w+', job_text)
            salary = ', '.join(salary_keywords)
            salary = salary.replace('K', '') 
        except:
            salary = ''        
        
        
        try:
            promoted = job_html.find(class_='job-card-container__footer-item inline-flex align-items-center').text
        except:
            promoted = ''
          
            
            
        new_row = pd.DataFrame({'job_titles': [title], 'companies': [company], 'job_links': [link], 'profile_matches':[profile_match_element],
                                'salary':[salary],'promoted':[promoted],'location':[location]})
        job_data = pd.concat([job_data, new_row], ignore_index=True)
        
        if(i/25) % 10 == 0:
            job_data.to_csv('jobs_data_science_remote.csv', index=False) # in each 10 pages, it is autosaving the df
        
    print(f'page {i/25} completed')
          
end = datetime.now()

print(f'Runtime of the program is: {end-start}')
#driver.close()

job_data.to_csv('jobs_data_science_remote.csv', index=False) # final save
```
