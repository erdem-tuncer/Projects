```python
start = datetime.now()

df_laptops = pd.DataFrame(columns=['title','quantity_sold', 'price', 'shipping_method','free_return','link'])

total_pages = 3

for i in range(1,total_pages+1):
    #print(i)
    driver.get('https://www.ebay.com/sch/i.html?_fsrp=1&_from=R40&RAM%2520Size=32%2520GB%7C64%2520GB&_nkw=gaming+laptop\
    &_sacat=175672&LH_BIN=1&Storage%2520Type=SSD%2520%2528Solid%2520State%2520Drive%2529&Processor=Intel%2520Core%2520i\
    7%252012th%2520Gen%252E%7CIntel%2520Core%2520i7%252013th%2520Gen%252E%7CIntel%2520Core%2520i9%252010th%2520Gen%252E%7\
    CIntel%2520Core%2520i9%252013th%2520Gen%252E&Screen%2520Size=15%252D15%252E9%2520in&_dcat=177&LH_PrefLoc=1&rt=nc&LH_FS=1\
    &imm=1&_pgn='
               +str(i))
    driver.maximize_window()
    time.sleep(np.random.randint(3, 6)) 
    
    #laptops = driver.find_element("ul[class*='results']>li[class*='bottom']")
    laptops = driver.find_elements(by="css selector", value="ul[class*='results']>li[class*='bottom']")
    
    
    for laptop in laptops:
        laptop_html = bs(laptop.get_attribute('innerHTML'), 'html.parser')
        
        try:
            title = laptop_html.find(class_='s-item__title').text
        except:
            title = ''
        
        try:
            quantity = laptop_html.find(class_='s-item__dynamic s-item__quantitySold').text
        except:
            quantity = ''
        
        try:
            price = laptop_html.find(class_='s-item__price').text.replace('$', '').replace(',', '')
        except:
            price = ''
        
        try:
            shipping = laptop_html.find(class_='s-item__shipping').text.replace('+$','').replace(',','').replace('shipping estimate', '')
        except:
            shipping =''
            
        try:
            free_return = laptop_html.find(class_='s-item__free-returns').text
        except:
            free_return = ''
        
        try:
            link = laptop_html.a['href']
        except:
            link = ''
            
        new_row = pd.DataFrame({'title': [title], 'quantity_sold': [quantity], 'price': [price], 'shipping_method':[shipping],
                                'free_return':[free_return], 'link':[link]})
        df_laptops = pd.concat([df_laptops, new_row], ignore_index=True)
        
        if(i) % 100 == 0:
            df_laptops.to_excel('gaming_laptops_ebay_us.xlsx', index=False) # in each 100 pages, it is autosaving the df
        
    print(f'page {i} completed')
          
end = datetime.now()

print(f'Runtime of the program is: {end-start}')
#driver.close()

df_laptops.to_excel('gaming_laptops_ebay_us.xlsx', index=False) # final save
```

```python
import pandas as pd
import re

# Initialize empty lists to store extracted specifications
ram_list = []
cpu_list = []
gpu_list = []
brand_list = []

# Define regular expressions to match patterns
ram_pattern = r'(\d+GB) RAM'
cpu_pattern = r'Intel Core i\d-\d+[\w-]*'
gpu_pattern = r'NVIDIA|GeForce|AMD'
brand_pattern = r'Razer|ACEMAGIC|Dell|HP|MSI|Acer|Lenovo|ASUS|Toshiba|GIGABYTE|Eluktronics|CHUWI|Gateway|PowerSpec'

# Loop through each row in the DataFrame
for title in df_laptops['title']:
    # Extract RAM
    ram_match = re.search(ram_pattern, title)
    if ram_match:
        ram_list.append(ram_match.group(1))
    else:
        ram_list.append(None)
    
    # Extract CPU
    cpu_match = re.search(cpu_pattern, title)
    if cpu_match:
        cpu_list.append(cpu_match.group())
    else:
        cpu_list.append(None)
    
    # Extract GPU
    gpu_match = re.search(gpu_pattern, title)
    if gpu_match:
        gpu_list.append(gpu_match.group())
    else:
        gpu_list.append(None)
    
    # Extract Brand
    brand_match = re.search(brand_pattern, title)
    if brand_match:
        brand_list.append(brand_match.group())
    else:
        brand_list.append(None)

# Add the extracted specifications as new columns in the DataFrame

df_laptops['RAM'] = ram_list
df_laptops['CPU'] = cpu_list
df_laptops['GPU'] = gpu_list
df_laptops['Brand'] = brand_list




# Define a function to extract RAM information from the title
def extract_ram(title):
    title = title.lower()  # Convert title to lowercase

    ram_match = re.search(r'(\d+)\s*g', title)
    ram_match2 = re.search(r'(\d+)\s*gb', title)
    
    if ram_match:
        return ram_match.group(1)
    elif ram_match2:
        return ram_match2.group(1)
    else:
        return None

# Apply the function to the 'title' column to create a new 'Extracted RAM' column
df_laptops['Extracted RAM'] = df_laptops['title'].apply(extract_ram)


# Fill the 'RAM' column with values from 'Extracted RAM' where 'RAM' is null
df_laptops['RAM'].fillna(df_laptops['Extracted RAM'], inplace=True)

# Drop the 'Extracted RAM' column if no longer needed
df_laptops.drop(columns=['Extracted RAM'], inplace=True)





# Display the updated DataFrame
df_laptops[['title', 'RAM', 'CPU', 'GPU', 'Brand']]
```
