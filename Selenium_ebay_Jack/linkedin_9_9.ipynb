```python
from datetime import datetime
import time
import numpy as np 
from bs4 import BeautifulSoup as bs


import re
from selenium import webdriver
import time


start = datetime.now()

job_data = pd.DataFrame(columns=['job_titles', 'companies', 'job_links', 'profile_matches', 'salary', 'promoted', 'location'])

total_pages = 3

for i in range(25, total_pages * 25 + 25, 25):
    url = 'https://www.linkedin.com/jobs/search/?currentJobId=3710744018&f_TPR=r604800&f_WT=2&geoId=103644278&keywords=data%20scientist&location=United%20States&refresh=true&start=' + str(i)
    driver.get(url)
    time.sleep(np.random.randint(9, 14))
    # Execute the JavaScript code to remove "Promoted" job listings
    javascript_code = '''
    var i = 0, x = document.getElementsByClassName('jobs-search-results-list')[0];
    !function e() {
        x.scroll(0, 100 * ++i);
        i > 40 ? null : setTimeout(e, 0);
    }();

    '''
    driver.execute_script(javascript_code)

    # Sleep for a random duration
    time.sleep(np.random.randint(12, 30))

    job_listings = driver.find_elements(by="css selector", value=".job-card-container")

    for job in job_listings:
        job_html = bs(job.get_attribute('innerHTML'), 'html.parser')

        try:
            title = job_html.find(class_='job-card-list__title').text
        except:
            title = ''

        try:
            company = job_html.find(class_='job-card-container__primary-description').text
        except:
            company = ''

        try:
            link = job_html.a['href']
        except:
            link = ''

        try:
            profile_match_element = job_html.find(class_='job-card-container__job-insight-text').text
        except:
            profile_match_element = ''

        try:
            location = job_html.find(class_='job-card-container__metadata-item').text
        except:
            location = ''

        try:
            job_text = job_html.get_text()
            salary_keywords = re.findall(r'\$\d+\s*[-–]\s*\$\d+|\$\d+(?:[\s\-–]\d+)?\s*(?:per|\/)?\s*\w+', job_text)
            salary = ', '.join(salary_keywords)
            salary = salary.replace('K', '') 
        except:
            salary = ''        

        try:
            promoted = job_html.find(class_='job-card-container__footer-item inline-flex align-items-center').text
        except:
            promoted = ''
          
        new_row = pd.DataFrame({'job_titles': [title], 'companies': [company], 'job_links': [link],
                                'profile_matches': [profile_match_element], 'salary': [salary],
                                'promoted': [promoted], 'location': [location]})
        job_data = pd.concat([job_data, new_row], ignore_index=True)

        if (i / 25) % 10 == 0:
            job_data.to_excel('jobs_data_science_remote.xlsx', index=False)

    print(f'page {i / 25} completed')

end = datetime.now()





print(f'Runtime of the program is: {end-start}')
#driver.close()
job_data['job_titles'] = job_data['job_titles'].str.strip()
job_data['companies'] = job_data['companies'].str.strip()
job_data['profile_matches'] = job_data['profile_matches'].str.strip()
job_data['salary'] = job_data['salary'].str.strip()
job_data['promoted'] = job_data['promoted'].str.strip()

job_data['job_links'] = r'https://www.linkedin.com/' + job_data['job_links']

job_data.to_excel('jobs_data_science_remote.xlsx',  index=False) # final save


# Close the WebDriver
#driver.quit()

```

```python
import pandas as pd


# Extract minimum and maximum salary from the salary column
job_data[['min_salary', 'max_salary']] = job_data['salary'].str.extract(r'\$?(\d+)(?:,\s*\$?(\d+))?')

# Convert the extracted values to numeric (remove ',' and '/hr' if present)
job_data['min_salary'] = pd.to_numeric(job_data['min_salary'].str.replace(',', '').str.replace('/hr', ''), errors='coerce')
job_data['max_salary'] = pd.to_numeric(job_data['max_salary'].str.replace(',', '').str.replace('/hr', ''), errors='coerce')

# Fill NaN values with 0 or any other appropriate value
job_data['min_salary'].fillna(0, inplace=True)
job_data['max_salary'].fillna(0, inplace=True)

job_data[['min_salary', 'max_salary']]
```
